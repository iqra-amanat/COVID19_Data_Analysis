```python
# COVID19 Data Analysis
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
data = pd.read_csv('C:\\Users\\Administrator\\Downloads\\country_wise_latest (2).csv')
data.fillna(0, inplace=True)  # Fill missing values

# Summarize Global Totals
global_totals = data[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum()
print("Global Totals:\n", global_totals)

# Top 10 Countries by Confirmed Cases
country_data = data.groupby('Country')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum().sort_values(by='Confirmed', ascending=False).head(10)

# Plot Top 10 Countries by Confirmed Cases
plt.figure(figsize=(12, 6))
sns.barplot(x=country_data.index, y=country_data['Confirmed'])
plt.title('Top 10 Countries by Confirmed COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.show()


# Mortality and Recovery Rates by Country
country_data['Mortality Rate (%)'] = (country_data['Deaths'] / country_data['Confirmed']) * 100
country_data['Recovery Rate (%)'] = (country_data['Recovered'] / country_data['Confirmed']) * 100

# Plot Mortality Rate
plt.figure(figsize=(12, 6))
sns.barplot(x=country_data.index, y=country_data['Mortality Rate (%)'], color='red')
plt.title('Top 10 Countries by Mortality Rate')
plt.xlabel('Country')
plt.ylabel('Mortality Rate (%)')
plt.xticks(rotation=45)
plt.show()

# Plot Recovery Rate
plt.figure(figsize=(12, 6))
sns.barplot(x=country_data.index, y=country_data['Recovery Rate (%)'], color='green')
plt.title('Top 10 Countries by Recovery Rate')
plt.xlabel('Country')
plt.ylabel('Recovery Rate (%)')
plt.xticks(rotation=45)
plt.show()

# Display Active Cases for Top 10 Countries
plt.figure(figsize=(12, 6))
sns.barplot(x=country_data.index, y=country_data['Active'], color='orange')
plt.title('Top 10 Countries by Active COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Active Cases')
plt.xticks(rotation=45)
plt.show()
country_data
```
